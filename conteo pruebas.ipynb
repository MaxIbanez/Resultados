```python
i = 0
tipo = "PCD"
iteraciones = "20000"
carpeta = iteraciones+"-"+tipo
vp = []
vn = []
fn = []
while i < 10:
    prueba = str(i+1)+"_resultado"
    archivo = iteraciones+"_"+str(i+1)+"_"+tipo+".txt"
    path = carpeta+"/"+prueba
    arc = open(path+"/"+archivo,"r")
    arc2 = open(path+"/resultados.txt","r")
    verdadero = 0
    for item in arc2:
        if len(item)>1:
            verdadero+=1
    print("="*20)
    print("prueba-> "+str(i+1))
    print("verdaderos:",verdadero)
    creados = 0
    falsos = 0
    for item in arc:
        if len(item)>1:
            creados+=1
        else:
            falsos+=1
    vp.append(verdadero)
    vn.append(creados-3-verdadero)
    fn.append(falsos)
    print("Creados:",creados-3-verdadero,"falsos:",falsos,"\ntotales:",(creados-3)+falsos)
    arc.close()
    arc2.close()
    i+=1
total = sum(vp)+sum(vn)+sum(fn)
print("VP:",sum(vp),sum(vp)*100/total,"VN:",sum(vn),sum(vn)*100/total,"FN:",sum(fn),sum(fn)*100/total)
```
